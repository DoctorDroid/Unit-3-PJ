```
input_ailments = ['Stress','Pain', 'Depression', 'Inflammation', 'Insomnia'] #, 'Lack of Appetite', 'Muscle Spasms', 'Nausea', 'Seizures' ]) #sample input recieved from web team
input_effects = ['Happy', 'Uplifted', 'Relaxed', 'Energetic', 'Creative',]# 'Euphoric', 'Sleepy', 'Hungry',
#                 'Focused', 'Aroused', 'Talkative', 'Giggly', 'Tingly', 'Dry Mouth', 'Horny', 'Paranoid', 'Anxious']
```

```
user_ailments = pd.DataFrame(input_ailments, columns = ['ail_type'])
```

```
user_ailments
```

```
user_effects = pd.DataFrame(input_effects, columns = ['effect_type'])
```

```
MJ= marijuana
MJ['fitness'] = 0
for i in range(len(MJ)):
    for ailment in user_ailments['ail_type']:
        if ailment in MJ['Ailment'][i]:
            MJ['fitness'][i] +=1
    for effect in user_effects['effect_type']:
        if effect in MJ['Effects_x'][i]:
            MJ['fitness'][i] +=1
        if effect in MJ['Effects_y'][i]:
            MJ['fitness'][i] +=1
```
